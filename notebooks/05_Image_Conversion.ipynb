Data Cleaning

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data
markov_path = '/content/qci_nist_results_nonrandom_part1 (0-149).csv'
qrng_path = '/content/qci_random_nist_window_results.csv'

qrng_df = pd.read_csv(qrng_path)
print(f"QRNG windows loaded: {len(qrng_df)}")

markov_df = pd.read_csv(markov_path)
print(f"Markov windows loaded: {len(markov_df)}")
```

```python
qrng_df['FrequencyTest'].value_counts()
markov_df['FrequencyTest'].value_counts()
```

```python
qrng_df.head()
```

```python
qrng_df['sourceLabel'] = 'Random'
```

```python
markov_df['sourceLabel'] = 'Non-Random'
```

```python
mapping = {' np.True_)': True, ' np.False_)': False, ' True)': True, ' False)': False}
```

```python
mapping[qrng_df['FrequencyTest'][0].split(',')[1]]
mapping[markov_df['FrequencyTest'][0].split(',')[1]]
```

```python
for col in qrng_df.columns:
  if qrng_df[col].dtype == 'object' and qrng_df[col].str.contains('np.True_\)|np.False_\)').any():
    qrng_df[col] = qrng_df[col].apply(lambda x: mapping[x.split(',')[1]])
```

```python
for col in markov_df.columns:
  if markov_df[col].dtype == 'object' and markov_df[col].str.contains('np.True_\)|np.False_\)').any():
    markov_df[col] = markov_df[col].apply(lambda x: mapping[x.split(',')[1]])
```

```python
mapping[qrng_df['MatrixRank'][0].split(',')[1]]
```

```python
mapping[markov_df['MatrixRank'][0].split(',')[1]]
```

```python
qrng_df['MatrixRank'] = qrng_df['MatrixRank'].apply(lambda x: mapping[x.split(',')[1]])
```

```python
markov_df['MatrixRank'] = markov_df['MatrixRank'].apply(lambda x: mapping[x.split(',')[1]])
```

```python
qrng_df.to_csv("final_random_cleaned.csv", index=False)
```

```python
markov_df.to_csv("final_nonrandom_cleaned.csv", index=False)
```

# Labelling

```python
cleaned_qrng = pd.read_csv("/content/final_random_cleaned.csv")
cleaned_markov = pd.read_csv("/content/final_nonrandom_cleaned.csv")
```

```python
cleaned_qrng.head()
```

```python
cleaned_qrng['Cumulative Sums'] = cleaned_qrng['Cumulative Sums (Forward)'] & cleaned_qrng['Cumulative Sums (Backward)']
```

```python
cleaned_markov.head()
```

```python
# List of NIST test columns
nist_tests = [
    'FrequencyTest',
    'BlockFrequency',
    'RunTest',
    'LongestRun',
    'MatrixRank',
    'Spectral',
    'Universal',
    'ApproxEntropy',
    'LinearComplexity',
    'Non-overlapping Template Matching Test',
    'Overlappong Template Matching Test',
    'Serial Test',
    'Cumulative Sums',
    'Random Excursion Test',
    'Random Excursion Variant Test'
]

# Count how many of these tests are True for each row
cleaned_qrng['nist_passing'] = cleaned_qrng[nist_tests].sum(axis=1)

# Create a "Critical tests" column — True if all 3 pass (FrequencyTest, RunTest, Spectral)
cleaned_qrng['Critical tests'] = (
    cleaned_qrng['FrequencyTest'] &
    cleaned_qrng['RunTest'] &
    cleaned_qrng['Spectral']
)
```

```python
# List of NIST test columns
nist_tests = [
    'FrequencyTest',
    'BlockFrequency',
    'RunTest',
    'LongestRun',
    'MatrixRank',
    'Spectral',
    'Universal',
    'ApproxEntropy',
    'LinearComplexity',
    'Non-overlapping Template Matching Test',
    'Overlappong Template Matching Test',
    'Serial Test',
    'Cumulative Sums',
    'Random Excursion Test',
    'Random Excursion Variant Test'
]

# Count how many of these tests are True for each row
cleaned_markov['nist_passing'] = cleaned_markov[nist_tests].sum(axis=1)

# Create a "Critical tests" column — True if all 3 pass (FrequencyTest, RunTest, Spectral)
cleaned_markov['Critical tests'] = (
    cleaned_markov['FrequencyTest'] &
    cleaned_markov['RunTest'] &
    cleaned_markov['Spectral']
)
```
